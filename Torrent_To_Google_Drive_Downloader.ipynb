Download Torrent or Magnet

```python
def download_torrent(magnet_or_torrent, download_path="/content/drive/MyDrive/Torrent_Downloads", is_magnet=True):
    ses = lt.session()
    ses.listen_on(6881, 6891)

    params = {
        'save_path': download_path,
        'storage_mode': lt.storage_mode_t.storage_mode_sparse
    }

    if is_magnet:
        print("🔗 Adding magnet link...")
        handle = lt.add_magnet_uri(ses, magnet_or_torrent, params)
    else:
        print("🔗 Adding torrent file...")
        info = lt.torrent_info(magnet_or_torrent)
        handle = ses.add_torrent({'ti': info, 'save_path': download_path})

    print("📥 Starting download...")
    while not handle.has_metadata():
        print("🔎 Fetching metadata...")
        time.sleep(1)

    print(f"📥 Downloading: {handle.name()}")
    while not handle.is_seed():
        s = handle.status()
        print(f"\r⏳ Progress: {s.progress * 100:.2f}%", end="")
        time.sleep(1)

    print(f"\n✅ Download complete: {handle.name()}")
    return os.path.join(download_path, handle.name())
```

Run the Downloader\
📌 Enter magnet link when asked\
📌 If error occurs check magnet url using
https://checker.openwebtorrent.com/ tool

```python
magnet = input("Enter Magnet Link or Torrent File Path:\n")
downloaded_path = download_torrent(magnet, is_magnet=magnet.startswith("magnet:?"))
```
